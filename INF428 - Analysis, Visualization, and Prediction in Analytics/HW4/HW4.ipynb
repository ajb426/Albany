```python
#data
df13 = data[(data.education_num == 13)]
df13 = df13.head(100)
df14 = data[(data.education_num == 14)]
df14 = df14.head(100)
df15 = data[(data.education_num == 15)]
df15 = df15.head(100)
df16 = data[(data.education_num == 16)]
df16 = df16.head(100)

```

```python
#lineplot
plt.plot(df13.capital_gain, df13.occupation, '-g', label = 'Bachelors')
plt.plot(df14.capital_gain, df14.occupation, ':b', label = 'Masters')
plt.plot(df15.capital_gain, df15.occupation, '-.k', label = 'Prof-school')
plt.plot(df16.capital_gain, df16.occupation, '--r', label = 'Doctorate')
plt.title("Capital Gain vs. Occupation by Education Level")
plt.xlabel("Capital Gain")
plt.ylabel("Occupation")
plt.xlim(0, 35000)
plt.legend()
```

```python
#scatterplot
plt.plot(df13.capital_gain, df13.occupation, 'ok', label='Bachelors')
plt.plot(df14.capital_gain, df14.occupation, '<g', label = 'Masters')
plt.plot(df15.capital_gain, df15.occupation, 'xb', label = 'Prof-School')
plt.plot(df16.capital_gain, df16.occupation, 'sr', label = 'Doctorate')
plt.title("Capital Gain vs. Occupation by Education Level")
plt.xlabel("Capital Gain")
plt.ylabel("Occupation")
plt.xlim(0,35000)
plt.legend()
```

```python
#boxplot data
box = data[['capital_gain']]
box.head(100)
box.describe()
box_array = np.array(box)
box_array= box_array[box_array!=0]
box_array
```

```python
#boxplot
colors = ['blue']
bp = plt.boxplot(box_array, patch_artist = True, notch = True)

for i in range(len(bp['boxes'])):
    bp['boxes'][i].set(facecolor=colors[i])
    bp['caps'][2*i + 1].set(color=colors[i])
    
plt.xticks([1], ['Capital Gain'])

plt.show()
```

```python
corr = data.corr()
ax = sns.heatmap(
    corr,
    vmin=-1,vmax=1, center=0,
    cmap=sns.diverging_palette(20, 220, n=200),
    square=True
)
ax.set_xticklabels(
ax.get_xticklabels(),
rotation=45,
horizontalalignment='right')
```

```python
comment_words = ''
stopwords = set(STOPWORDS)
for val in data.native_country:
    val = str(val)
    tokens = val.split()
    for i in range(len(tokens)):
        tokens[i] = tokens[i].lower()
        comment_words += " ".join(tokens)+" "

        
wordcloud = WordCloud(width = 800, height = 800,
                     background_color = 'white',
                     stopwords = stopwords,
                     min_font_size = 10).generate(comment_words)
plt.figure(figsize = (8,8), facecolor = None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```
